```python
#Given a DataFrame df, how do you check the data type of each column and find out how many missing (null) values exist in each column?
```

```python
import pandas as pd

df = pd.DataFrame({
    'name':  ['Alice', 'Bob', None, 'David'],
    'age':   [25, None, 30, 22],
    'score': [88.5, 91.0, 76.5, None]
})

print(df.dtypes)

print(df.isnull().sum())
```
