```python

#Sports Covered: Football, Basketball, Cricket, Tennis, Athletics


import pandas as pd
import numpy as np
from io import StringIO


#FOOTBALL 
football_data = """Player,Country,Goals,Assists,Matches,Position
Lionel Messi,Argentina,13,8,26,Forward
Cristiano Ronaldo,Portugal,9,3,22,Forward
Kylian Mbappe,France,12,5,23,Forward
Kevin De Bruyne,Belgium,4,11,22,Midfielder
Robert Lewandowski,Poland,10,4,20,Forward
Harry Kane,England,11,3,24,Forward
Vinicius Jr,Brazil,9,6,20,Forward
Pedri,Spain,2,6,18,Midfielder
Bukayo Saka,England,5,6,22,Midfielder
Antoine Griezmann,France,7,8,23,Forward
"""

# BASKETBALL (NBA)
basketball_data = """Player,Team,Points_Per_Game,Rebounds,Assists,Steals,FG_Percent
LeBron James,Lakers,27.1,7.3,8.1,1.3,52.4
Stephen Curry,Warriors,30.1,5.2,6.3,1.5,49.3
Kevin Durant,Suns,29.1,6.7,5.0,0.9,53.7
Giannis Antetokounmpo,Bucks,31.1,11.8,5.7,0.8,55.3
Joel Embiid,76ers,33.1,10.2,4.2,1.0,52.8
Nikola Jokic,Nuggets,26.4,12.4,9.0,1.3,63.2
Luka Doncic,Mavericks,32.4,8.6,8.0,1.4,49.6
Jayson Tatum,Celtics,26.9,8.1,4.9,1.0,47.3
Damian Lillard,Bucks,24.3,4.4,7.3,0.9,43.2
Anthony Davis,Lakers,24.7,12.6,3.5,1.2,56.2
"""

#CRICKET (IPL)
cricket_data = """Player,Team,Runs,Wickets,Matches,Average,Strike_Rate
Virat Kohli,RCB,741,0,16,61.75,139.5
Rohit Sharma,MI,502,0,14,38.61,131.2
MS Dhoni,CSK,236,0,15,47.20,168.3
Jasprit Bumrah,MI,0,20,14,18.45,0.0
Ravindra Jadeja,CSK,180,15,15,36.00,142.6
KL Rahul,LSG,616,0,15,51.33,135.8
Shubman Gill,GT,890,0,17,59.33,157.9
Mohammed Shami,GT,0,28,17,19.21,0.0
Suryakumar Yadav,MI,355,0,14,29.58,169.4
Hardik Pandya,GT,344,11,15,34.40,131.0
"""

#TENNIS
tennis_data = """Player,Country,Grand_Slams,Win_Rate,Ranking,Aces_Per_Match,Age
Novak Djokovic,Serbia,24,85.2,1,6.2,36
Carlos Alcaraz,Spain,2,80.4,2,5.8,20
Daniil Medvedev,Russia,1,76.3,3,8.1,27
Jannik Sinner,Italy,1,78.6,4,7.3,22
Alexander Zverev,Germany,0,72.5,5,9.2,26
Andrey Rublev,Russia,0,68.4,6,6.7,25
Holger Rune,Denmark,0,65.2,7,7.5,20
Stefanos Tsitsipas,Greece,0,70.1,8,6.9,25
Rafael Nadal,Spain,22,83.5,9,3.8,37
Roger Federer,Switzerland,20,82.1,10,8.6,42
"""

#ATHLETICS 
athletics_data = """Athlete,Country,Best_Time,Medal,Year,Age
Usain Bolt,Jamaica,9.58,Gold,2009,22
Tyson Gay,USA,9.69,Silver,2009,26
Asafa Powell,Jamaica,9.72,Bronze,2008,25
Justin Gatlin,USA,9.74,Gold,2015,33
Christian Coleman,USA,9.76,Gold,2019,23
Marcell Jacobs,Italy,9.80,Gold,2021,26
Yohan Blake,Jamaica,9.69,Silver,2012,22
Andre De Grasse,Canada,9.89,Bronze,2021,26
Akani Simbine,South Africa,9.84,Silver,2022,28
Trayvon Bromell,USA,9.77,Bronze,2021,25
"""


# LOAD DATAFRAMES


df_football    = pd.read_csv(StringIO(football_data))
df_basketball  = pd.read_csv(StringIO(basketball_data))
df_cricket     = pd.read_csv(StringIO(cricket_data))
df_tennis      = pd.read_csv(StringIO(tennis_data))
df_athletics   = pd.read_csv(StringIO(athletics_data))


# ANALYSIS FUNCTIONS

def section(title):
    print("\n" + "═" * 55)
    print(f"  {title}")
    print("═" * 55)

def show_table(df, cols, title=""):
    if title:
        print(f"\n  [{title}]")
    print(df[cols].to_string(index=False))
```

```python
# 1. FOOTBALL ANALYSIS

section("⚽ FOOTBALL — FIFA World Cup Stats")

df_football["Goal_Contribution"] = df_football["Goals"] + df_football["Assists"]
df_football["Goals_Per_Match"]   = (df_football["Goals"] / df_football["Matches"]).round(2)

print(f"\n  Total Goals     : {df_football['Goals'].sum()}")
print(f"  Total Assists   : {df_football['Assists'].sum()}")
print(f"  Avg Goals/Match : {df_football['Goals_Per_Match'].mean():.2f}")

print("\n  Top 5 by Goal Contribution:")
top5 = df_football.nlargest(5, "Goal_Contribution")[["Player","Goals","Assists","Goal_Contribution"]]
print(top5.to_string(index=False))

print("\n  Position-wise avg goals:")
print(df_football.groupby("Position")["Goals"].mean().round(2).to_string())

print("\n  Most efficient scorer (Goals/Match):")
best = df_football.loc[df_football["Goals_Per_Match"].idxmax()]
print(f"  → {best['Player']} — {best['Goals_Per_Match']} goals/match")
```

```python
# 2. BASKETBALL ANALYSIS

section("🏀 BASKETBALL — NBA Player Stats")

df_basketball["Impact_Score"] = (
    df_basketball["Points_Per_Game"] * 1.0 +
    df_basketball["Rebounds"]        * 0.8 +
    df_basketball["Assists"]         * 0.7 +
    df_basketball["Steals"]          * 1.5
).round(2)

print(f"\n  Avg Points/Game : {df_basketball['Points_Per_Game'].mean():.1f}")
print(f"  Avg Rebounds    : {df_basketball['Rebounds'].mean():.1f}")
print(f"  Avg Assists     : {df_basketball['Assists'].mean():.1f}")
print(f"  Best FG%        : {df_basketball['FG_Percent'].max()}%  ({df_basketball.loc[df_basketball['FG_Percent'].idxmax(), 'Player']})")

print("\n  Top 5 by Impact Score:")
top_bball = df_basketball.nlargest(5, "Impact_Score")[["Player","Points_Per_Game","Rebounds","Assists","Impact_Score"]]
print(top_bball.to_string(index=False))

print("\n  Correlation — Points vs FG%:")
corr = df_basketball["Points_Per_Game"].corr(df_basketball["FG_Percent"])
print(f"  → Pearson r = {corr:.3f}  ({'positive' if corr > 0 else 'negative'} relationship)")
```

```python
# 3. CRICKET ANALYSIS

section("🏏 CRICKET — IPL Season Stats")

batsmen  = df_cricket[df_cricket["Runs"] > 0].copy()
bowlers  = df_cricket[df_cricket["Wickets"] > 0].copy()

batsmen["Runs_Per_Match"]    = (batsmen["Runs"] / batsmen["Matches"]).round(2)
bowlers["Wickets_Per_Match"] = (bowlers["Wickets"] / bowlers["Matches"]).round(2)

print(f"\n  Total Runs Scored   : {df_cricket['Runs'].sum()}")
print(f"  Total Wickets Taken : {df_cricket['Wickets'].sum()}")
print(f"  Avg Batting Average : {batsmen['Average'].mean():.2f}")
print(f"  Avg Strike Rate     : {batsmen['Strike_Rate'].mean():.1f}")

print("\n  Top 5 Run Scorers:")
show_table(batsmen.nlargest(5, "Runs"), ["Player","Team","Runs","Average","Strike_Rate"])

print("\n  Top 3 Wicket Takers:")
show_table(bowlers.nlargest(3, "Wickets"), ["Player","Team","Wickets","Wickets_Per_Match"])
```

```python
# 4. TENNIS ANALYSIS

section(" TENNIS — ATP Player Stats")

df_tennis["Slam_Ratio"] = (df_tennis["Grand_Slams"] / df_tennis["Age"] * 10).round(2)

print(f"\n  Total Grand Slams (top 10)  : {df_tennis['Grand_Slams'].sum()}")
print(f"  Avg Win Rate                : {df_tennis['Win_Rate'].mean():.1f}%")
print(f"  Avg Aces Per Match          : {df_tennis['Aces_Per_Match'].mean():.1f}")

print("\n  Top 5 by Win Rate:")
show_table(df_tennis.nlargest(5, "Win_Rate"), ["Player","Country","Grand_Slams","Win_Rate","Ranking"])

print("\n  Youngest vs Oldest:")
youngest = df_tennis.loc[df_tennis["Age"].idxmin()]
oldest   = df_tennis.loc[df_tennis["Age"].idxmax()]
print(f"  → Youngest: {youngest['Player']} ({youngest['Age']} yrs) — Win Rate {youngest['Win_Rate']}%")
print(f"  → Oldest  : {oldest['Player']} ({oldest['Age']} yrs) — Win Rate {oldest['Win_Rate']}%")

print("\n  Age vs Win Rate (binned):")
df_tennis["Age_Group"] = pd.cut(df_tennis["Age"], bins=[19,25,30,45], labels=["20-25","26-30","31+"])
print(df_tennis.groupby("Age_Group", observed=True)["Win_Rate"].mean().round(2).to_string())
```

```python
# 5. ATHLETICS ANALYSIS

section(" ATHLETICS — 100m Sprint Records")

df_athletics["Time_Rank"] = df_athletics["Best_Time"].rank().astype(int)

print(f"\n  Fastest Time (World Record) : {df_athletics['Best_Time'].min()}s  ({df_athletics.loc[df_athletics['Best_Time'].idxmin(), 'Athlete']})")
print(f"  Average Best Time          : {df_athletics['Best_Time'].mean():.3f}s")
print(f"  Time Range (fastest–slowest): {df_athletics['Best_Time'].max() - df_athletics['Best_Time'].min():.2f}s")

print("\n  Ranked by Speed:")
show_table(df_athletics.sort_values("Best_Time")[["Athlete","Country","Best_Time","Medal","Year"]], 
           ["Athlete","Country","Best_Time","Medal","Year"])

print("\n  Gold medalists only:")
golds = df_athletics[df_athletics["Medal"] == "Gold"].sort_values("Best_Time")
show_table(golds, ["Athlete","Country","Best_Time","Year"])

print("\n  Country-wise athlete count:")
print(df_athletics["Country"].value_counts().to_string())
```

```python
# CROSS-SPORT SUMMARY
section("📊 CROSS-SPORT SUMMARY")

summary = pd.DataFrame({
    "Sport"      : ["Football","Basketball","Cricket","Tennis","Athletics"],
    "Players"    : [len(df_football), len(df_basketball), len(df_cricket), len(df_tennis), len(df_athletics)],
    "Key Metric" : ["Goals/Assists","Points Per Game","Runs/Wickets","Win Rate %","100m Time (s)"],
    "Top Performer": [
        df_football.loc[df_football["Goal_Contribution"].idxmax(), "Player"],
        df_basketball.loc[df_basketball["Impact_Score"].idxmax(), "Player"],
        df_cricket.loc[df_cricket["Runs"].idxmax(), "Player"],
        df_tennis.loc[df_tennis["Win_Rate"].idxmax(), "Player"],
        df_athletics.loc[df_athletics["Best_Time"].idxmin(), "Athlete"],
    ]
})

print("\n")
print(summary.to_string(index=False))
print("\n" + "═" * 55)
print("  ✅ Analysis Complete — 5 Sports | 50 Athletes")
print("═" * 55)
```
