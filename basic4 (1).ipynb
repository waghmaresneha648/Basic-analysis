```python
#Using the DataFrame below, do two things: (1) sort the data by score in descending order, and (2) print the mean, median, and standard deviation of the score column.
```

```python
import pandas as pd

df = pd.DataFrame({
    'name':  ['Alice', 'Bob', 'Carol', 'David', 'Eva'],
    'score': [88, 76, 95, 82, 91]
})

sorted_df = df.sort_values(by='score', ascending=False)
print(sorted_df)

print("\nMean:  ", df['score'].mean())
print("Median:", df['score'].median())
print("Std:   ", df['score'].std().round(2))
```
