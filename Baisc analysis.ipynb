```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("C:\\Users\\LENOVO\\Downloads\\data_analysis_practice (1).csv")
```

```python

df["total_amount"] = df["price"] * df["quantity"]


df.head()


df.groupby("city")["total_amount"].sum()


df.groupby("product")["price"].mean()


df.groupby("product")["quantity"].sum()


df.groupby("category")["total_amount"].sum().sort_values(ascending=False)

# Q27
df.groupby("city")["age"].mean()

# Q28
df.groupby("gender")["order_id"].count()

# Q29
df.groupby("product")["quantity"].sum().sort_values(ascending=False).head(3)

# Q30
df.groupby("city")["total_amount"].sum().idxmax()
```

```python
# Q31
df.sort_values(by="price", ascending=False)

# Q32
df.sort_values(by="date", ascending=False)

# Q33
product_revenue = df.groupby("product")["total_amount"].sum()
product_revenue.rank(ascending=False)

# Q34
df.sort_values(by="total_amount", ascending=False).head(1)

# Q35
df.sort_values(by="price").head(1)

# Q36
df["customer_id"].value_counts()[df["customer_id"].value_counts() > 1]

# Q37
df.groupby("date")["total_amount"].sum().sort_values(ascending=False)

# Q38
def price_category(p):
    if p < 5000:
        return "Low"
    elif p <= 20000:
        return "Medium"
    else:
        return "High"

df["price_category"] = df["price"].apply(price_category)

# Q39
category_sales = df.groupby("category")["total_amount"].sum()
(category_sales / category_sales.sum()) * 100

# Q40 (Outliers using IQR)
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df["price"] < Q1 - 1.5*IQR) | (df["price"] > Q3 + 1.5*IQR)]
outliers
```

```python
# Q41
df.groupby("city")["total_amount"].sum().plot(kind="bar")
plt.title("Total Sales per City")
plt.show()

# Q42
df.groupby("product")["total_amount"].sum().plot(kind="bar")
plt.title("Revenue per Product")
plt.show()

# Q43
df.groupby("category")["total_amount"].sum().plot(kind="pie", autopct='%1.1f%%')
plt.title("Category-wise Sales Distribution")
plt.ylabel("")
plt.show()

# Q44
df["age"].plot(kind="hist", bins=10)
plt.title("Age Distribution")
plt.show()

# Q45
df.groupby("date")["total_amount"].sum().plot(kind="line")
plt.title("Sales Trend Over Time")
plt.show()

# Q46
df.groupby("gender")["total_amount"].sum()

# Q47
df[["age", "total_amount"]].corr()

# Q48
df.groupby(["city", "product"])["quantity"].sum().reset_index().sort_values(["city","quantity"], ascending=False)

# Q49 (Customer Segmentation)
customer_spending = df.groupby("customer_id")["total_amount"].sum()

def segment(x):
    if x < 5000:
        return "Low Value"
    elif x < 30000:
        return "Mid Value"
    else:
        return "High Value"

customer_segment = customer_spending.apply(segment)
customer_segment.value_counts()

# Q50 (Insights)
df.groupby("product")["total_amount"].sum().sort_values(ascending=False)
```
