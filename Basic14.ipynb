```python
You have a wide-format DataFrame where each quarter is its own column. Do the following:

(1) Convert it to long format using pd.melt() so each row represents one region-quarter combination.
(2) Rename the melted columns to quarter and sales.
(3) Then find the top performing quarter (highest total sales across all regions).

```

```python
import pandas as pd

df = pd.DataFrame({
    'region': ['North', 'South', 'West'],
    'Q1':     [32000,   28000,  25000],
    'Q2':     [41000,   31000,  29000],
    'Q3':     [38000,   35000,  33000],
    'Q4':     [47000,   42000,  39000]
})

melted = pd.melt(
    df,
    id_vars='region',
    value_vars=['Q1','Q2','Q3','Q4'],
    var_name='quarter',
    value_name='sales'
)

print(melted)

quarter_totals = melted.groupby('quarter')['sales'].sum()
top_quarter = quarter_totals.idxmax()
print(f"\nQuarter totals:\n{quarter_totals}")
print(f"\nTop performing quarter: {top_quarter} "
      f"(₹{quarter_totals[top_quarter]:,})")
```
